```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
## connecting to google
trends = TrendReq()
```

```python
trends.build_payload(kw_list=["Cloud Computing"]) ## method to get the data for the mentioned keyword 
dataframe = trends.interest_by_region()  ## method returns data for where the keyword is most searched as shown on Google Trends' Interest by Region section.
data = dataframe.sort_values(by="Cloud Computing",ascending=False)
print(data)
```

```python
## Getting the top 10 countries who searched for "Cloud Computing"
data = data.head(10)
data
```

```python
data = data.reset_index()  ## reset_index method provides the indices to each rows as shown below 
data
```

```python
## Plotting a bar graph for the top 10 countries
data.plot(x="geoName",y="Cloud Computing",figsize=(20,15),kind="bar")   ## Pandas has a built in .plot() function as part of the DataFrame class.:
plt.style.use('fivethirtyeight')  ## this style.use method will put lines in the plain graph as shown below
plt.show()
```

```python
## Visualising total search queries based on â€œCloud Computingâ€ increased or decreased on Google:
totaldata = TrendReq(hl='en-US',tz=360)  ## hl is host language, tz is Timezone Offset (in minutes)= 360 for US 
totaldata.build_payload(kw_list=['Cloud Computing'])
total_data = totaldata.interest_over_time()  ## returns historical, indexed data for when the keyword was searched most as shown on Google Trends' Interest Over Time section.
total_data
```

```python
figure,axes = plt.subplots(figsize=(20,15))
total_data['Cloud Computing'].plot()
plt.style.use('fivethirtyeight')
plt.title("Total Google Searches for Cloud Computing",fontweight="bold")
plt.xlabel("YEAR")
plt.ylabel("TOTAL  COUNT")
plt.show()
```

## CONCLUSION::
    So we can see that searches based on â€œCloud Computingâ€ on Google started to increase in 2017 and the highest searches were done in starting of 2017 till today. This is how we can analyze Google searches based on any keyword. A business can perform Google search analysis to understand what people are looking for on Google at any given time.
